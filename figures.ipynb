```python
import ase.io
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from utils.energy_util import calc_pair_distances
```

```python
xe2 = 'dataset/xe2_50.xyz'
xe3 = 'dataset/xe3_50.xyz'
rand_xe3 = 'dataset/xe3_dataset_dft.xyz'
```

```python
dataset = ase.io.read(xe2, ":") + ase.io.read(xe3, ":") + ase.io.read(rand_xe3, ":")
```

# Atom pair distance distribution

```python
distances = np.concatenate([calc_pair_distances(atoms) for atoms in dataset], axis=0)
```

```python
sns.histplot(distances)
```
